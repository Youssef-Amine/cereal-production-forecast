There are two missing values ​​for production in 2011 and one for area in the same year. We will handle them by replacing them with the average of the values ​​in the row to which each missing value belongs.

```python
#Processing of the production base
# Imputation of NaNs in column '2011' by the average of the corresponding row
def impute_2011(row):
    if pd.isna(row['2011']):
        numeric_vals = row.drop(labels='2011')  # we exclude the '2011' column itself
        numeric_vals = pd.to_numeric(numeric_vals, errors='coerce')  # convert to numeric, ignore strings
        return numeric_vals.mean(skipna=True)
    else:
        return row['2011']

df_prod['2011'] = df_prod.apply(impute_2011, axis=1)
```

We have a maximum of 11 rows, whereas there are 13 regions in Burkina Faso. This means that some regions have no values ​​at all. Furthermore, there are missing values ​​for these dates. Taking a look at the database, we see, for example, that the Center-East and Center-North regions have no values ​​from 1996 to 2005. Here's how we'll handle the missing values:

* For regions that have at least a certain number of values, we'll perform an imputation using the average of the missing values.
* For regions that don't exist (Cascades and Center-West), we'll create rows for them and process them using a method similar to k-nearest neighbors.
Indeed, the Cascades region is closer to the Hauts-Bassins and Southwest regions. Thus, for each year, the number of rainy days/year corresponding to the Cascades region will be equal to the average of the values ​​for the Hauts-Bassins and the Southwest.
For the Center-West, we will consider the Center, Boucle du Mouhoun, and Center-South as neighbors.

* The Plateau Central region only has values ​​from 2015 to 2022. This is insufficient to perform an average imputation for all other missing values. Therefore, for these missing values, we will take the average of the values ​​for the Center and Center-East regions.




```python
#Processing missing values ​​from the number of rainy days database.
annees = [str(annee) for annee in range(1996, 2023)]

# 1. Imputation for Centre-East, Centre-North, Sahel: replace NaN year by year
for region in ['Centre-Est', 'Centre-Nord', 'Sahel']:
    idx = df_pluie[df_pluie['Région'] == region].index
    if not idx.empty:
        for annee in annees:
            valeur = df_pluie.loc[idx, annee].values[0]
            # If missing value, impute by the average of other non-missing years in the same region
            if pd.isna(valeur):
                moyenne_autres = df_pluie.loc[idx, annees].mean(axis=1, skipna=True).values[0]
                df_pluie.loc[idx, annee] = moyenne_autres

# 2. Imputation for Plateau Central: average of the values ​​for the Centre and Centre-East regions
idx_pc = df_pluie[df_pluie['Région'] == 'Plateau Central'].index
idx_centre = df_pluie[df_pluie['Région'] == 'Centre'].index
idx_centre_est = df_pluie[df_pluie['Région'] == 'Centre-Est'].index

if not idx_pc.empty and not idx_centre.empty and not idx_centre_est.empty:
    for annee in annees:
        valeur_centre = df_pluie.loc[idx_centre, annee].values[0]
        valeur_centre_est = df_pluie.loc[idx_centre_est, annee].values[0]
        moyenne = (valeur_centre + valeur_centre_est) / 2
        df_pluie.loc[idx_pc, annee] = moyenne

# 3. Create the Cascades modality: average of the Hauts-Bassins and South-West
idx_hauts_bassins = df_pluie[df_pluie['Région'] == 'Hauts-Bassins'].index
idx_sud_ouest = df_pluie[df_pluie['Région'] == 'Sud-Ouest'].index

if not idx_hauts_bassins.empty and not idx_sud_ouest.empty:
    nouvelles_valeurs = {}
    for annee in annees:
        val_hb = df_pluie.loc[idx_hauts_bassins, annee].values[0]
        val_so = df_pluie.loc[idx_sud_ouest, annee].values[0]
        nouvelles_valeurs[annee] = (val_hb + val_so) / 2
    nouvelles_valeurs['Région'] = 'Cascades'
    nouvelles_valeurs['Units'] = df_pluie['Units'].iloc[0]  # retake Units value
    df_pluie = pd.concat([df_pluie, pd.DataFrame([nouvelles_valeurs])], ignore_index=True)

# 4. Create the Centre-West modality: average of the Centre, Boucle du Mouhoun, Centre-Sud
idx_boucle = df_pluie[df_pluie['Région'] == 'Boucle du Mouhoun'].index
idx_centre_sud = df_pluie[df_pluie['Région'] == 'Centre-Sud'].index

if not (idx_centre.empty or idx_boucle.empty or idx_centre_sud.empty):
    nouvelles_valeurs = {}
    for annee in annees:
        val_centre = df_pluie.loc[idx_centre, annee].values[0]
        val_boucle = df_pluie.loc[idx_boucle, annee].values[0]
        val_cs = df_pluie.loc[idx_centre_sud, annee].values[0]
        nouvelles_valeurs[annee] = (val_centre + val_boucle + val_cs) / 3
    nouvelles_valeurs['Région'] = 'Centre-Ouest'
    nouvelles_valeurs['Units'] = df_pluie['Units'].iloc[0]
    df_pluie = pd.concat([df_pluie, pd.DataFrame([nouvelles_valeurs])], ignore_index=True)

# Show result
df_pluie.head()
```

Precipiations a exactement les mêmes caractéristiques que Nombre de jours de pluie en termes de valeurs manquantes. Donc c'est le même code qui sera appliqué.

```python
#Traitement valeurs manquantes de la base précipitations moyennes annuelles.
annees = [str(annee) for annee in range(1996, 2023)]

# 1. Imputation pour Centre-Est, Centre-Nord, Sahel : remplacer NaN par la moyenne de la ligne (en ignorant les NaN)
for region in ['Centre-Est', 'Centre-Nord', 'Sahel']:
    idx = df_precip[df_precip['Région'] == region].index
    if not idx.empty:
        moyenne_ligne = df_precip.loc[idx, annees].mean(axis=1, skipna=True)
        for annee in annees:
            df_precip.loc[idx, annee] = df_precip.loc[idx, annee].fillna(moyenne_ligne.values[0])

# 2. Imputation pour Plateau Central : moyenne des valeurs des régions Centre et Centre-Est
idx_pc = df_precip[df_precip['Région'] == 'Plateau Central'].index
idx_centre = df_precip[df_precip['Région'] == 'Centre'].index
idx_centre_est = df_precip[df_precip['Région'] == 'Centre-Est'].index

if not idx_pc.empty and not idx_centre.empty and not idx_centre_est.empty:
    for annee in annees:
        valeur_centre = df_precip.loc[idx_centre, annee].values[0]
        valeur_centre_est = df_precip.loc[idx_centre_est, annee].values[0]
        moyenne = (valeur_centre + valeur_centre_est) / 2
        if pd.isna(df_precip.loc[idx_pc, annee].values[0]):
            df_precip.loc[idx_pc, annee] = moyenne

# 3. Créer la modalité Cascades : moyenne des Hauts-Bassins et Sud-Ouest
idx_hauts_bassins = df_precip[df_precip['Région'] == 'Hauts-Bassins'].index
idx_sud_ouest = df_precip[df_precip['Région'] == 'Sud-Ouest'].index

if not idx_hauts_bassins.empty and not idx_sud_ouest.empty:
    nouvelles_valeurs = {}
    for annee in annees:
        val_hb = df_precip.loc[idx_hauts_bassins, annee].values[0]
        val_so = df_precip.loc[idx_sud_ouest, annee].values[0]
        nouvelles_valeurs[annee] = (val_hb + val_so) / 2
    nouvelles_valeurs['Région'] = 'Cascades'
    nouvelles_valeurs['Units'] = df_precip['Units'].iloc[0]  # reprendre la valeur Units
    df_precip = pd.concat([df_precip, pd.DataFrame([nouvelles_valeurs])], ignore_index=True)

# 4. Créer la modalité Centre-Ouest : moyenne des Centre, Boucle du Mouhoun, Centre-Sud
idx_boucle = df_precip[df_precip['Région'] == 'Boucle du Mouhoun'].index
idx_centre_sud = df_precip[df_precip['Région'] == 'Centre-Sud'].index

if not (idx_centre.empty or idx_boucle.empty or idx_centre_sud.empty):
    nouvelles_valeurs = {}
    for annee in annees:
        val_centre = df_precip.loc[idx_centre, annee].values[0]
        val_boucle = df_precip.loc[idx_boucle, annee].values[0]
        val_cs = df_precip.loc[idx_centre_sud, annee].values[0]
        nouvelles_valeurs[annee] = (val_centre + val_boucle + val_cs) / 3
    nouvelles_valeurs['Région'] = 'Centre-Ouest'
    nouvelles_valeurs['Units'] = df_pluie['Units'].iloc[0]
    df_precip = pd.concat([df_precip, pd.DataFrame([nouvelles_valeurs])], ignore_index=True)

# Afficher le résultat
df_precip.head()
```

Here, we have a maximum of 8 regions. The missing regions are: Centre-East, Centre-North, Centre-West, Cascades, and Plateau Central.

1. There are some missing values ​​for some regions, which will be processed by imputing the average of the values ​​for that region.
2. Missing regions will be processed by imputing the average of the values ​​for neighboring regions for each year.
Centre.
* Cascades: Average between Hauts-Bassins and Southwest
* Centre-West: Average between Boucle du Mouhoun, Centre, and Centre-South
* Centre-East: Average between Centre and East
* Centre-North: Average between Centre and North
* Plateau Central: Average between Centre and Centre-East







```python
#Traitement valeurs manquantes de la base vitesse moyenne du vent.
annees = [str(an) for an in range(1996, 2023)]

# Imputation des valeurs manquantes pour les régions présentes
for region in df_vent['Région'].unique():
    idx = df_vent[df_vent['Région'] == region].index
    for an in annees:
        if pd.isna(df_vent.loc[idx, an].values[0]):
            # Calculer la moyenne de la région sur les années disponibles
            moyenne_region = df_vent.loc[idx, annees].mean(axis=1).values[0]
            df_vent.loc[idx, an] = moyenne_region

# Créer les nouvelles modalités

# Fonction pour calculer la moyenne entre plusieurs régions
def moyenne_regions(df, regions_source):
    return df[df['Région'].isin(regions_source)][annees].mean()

# Cascades = Moyenne (Hauts-Bassins, Sud-Ouest)
valeurs_cascades = moyenne_regions(df_vent, ['Hauts-Bassins', 'Sud-Ouest'])
df_vent = pd.concat([df_vent, pd.DataFrame({'Région': ['Cascades'], 'Units': ['m/s'], **valeurs_cascades.to_dict()})], ignore_index=True)

# Centre-Ouest = Moyenne (Centre, Boucle du Mouhoun, Centre-Sud)
valeurs_centre_ouest = moyenne_regions(df_vent, ['Centre', 'Boucle du Mouhoun', 'Centre-Sud'])
df_vent = pd.concat([df_vent, pd.DataFrame({'Région': ['Centre-Ouest'], 'Units': ['m/s'], **valeurs_centre_ouest.to_dict()})], ignore_index=True)

# Centre-Est = Moyenne (Centre, Est)
valeurs_centre_est = moyenne_regions(df_vent, ['Centre', 'Est'])
df_vent = pd.concat([df_vent, pd.DataFrame({'Région': ['Centre-Est'], 'Units': ['m/s'], **valeurs_centre_est.to_dict()})], ignore_index=True)

# Centre-Nord = Moyenne (Centre, Nord)
valeurs_centre_nord = moyenne_regions(df_vent, ['Centre', 'Nord'])
df_vent = pd.concat([df_vent, pd.DataFrame({'Région': ['Centre-Nord'], 'Units': ['m/s'], **valeurs_centre_nord.to_dict()})], ignore_index=True)

# Plateau Central = Moyenne (Centre, Centre-Est)
valeurs_plateau_central = moyenne_regions(df_vent, ['Centre', 'Centre-Est'])
df_vent = pd.concat([df_vent, pd.DataFrame({'Région': ['Plateau Central'], 'Units': ['m/s'], **valeurs_plateau_central.to_dict()})], ignore_index=True)

# Visualiser le résultat
df_vent.head()
```

Now that the databases have been processed, we'll move on to merging them into a single database. For each database, we'll transform them by putting the years in a Year column, create a column for the information provided by the database (Production, Area, Average Annual Precipitation, Temperature, etc.), and sort by region and grain if there's a grain and year column.
Then the next transformed database will be merged with the previous one. After merging the Production and Area databases, we'll create a new Yield variable.

```python
# Transformation of the production base:
#Removal of unnecessary Burkina lines
df_prod = df_prod[df_prod['Région'] != 'Burkina Faso']

#owe convert the year columns into rows
df_prod_trans = pd.melt(df_prod,
                  id_vars=['Région', 'Céréales'],  # We keep these columns fixed
                  var_name='Année',
                  value_name='Production (en tonnes)')

df_prod_trans = df_prod_trans[df_prod_trans['Année'].str.isnumeric()]
# Convert Year to int
df_prod_trans['Année'] = df_prod_trans['Année'].astype(int)

# Sort: Region ➔ Type of cereal ➔ Year
df_prod_trans = df_prod_trans.sort_values(by=['Région', 'Céréales', 'Année'])

print(df_prod_trans.head())
```

```python
# Transformer la base Superficie
#Suppression des lignes Burkina qui ne sont pas necessaires
df_superf = df_superf[df_superf['Région'] != 'Burkina Faso']

#on convertit les colonnes années en lignes
df_superf_trans = pd.melt(df_superf,
                             id_vars=['Région', 'Céréales'],
                             var_name='Année',
                             value_name='Superficie (en ha)')
df_superf_trans = df_superf_trans[df_superf_trans['Année'].str.isnumeric()]
df_superf_trans = df_superf_trans.sort_values(by=['Région', 'Céréales', 'Année'])
df_superf_trans['Année'] = df_superf_trans['Année'].astype(int)
# Fusionner les deux bases
df_final1 = pd.merge(df_prod_trans,
                    df_superf_trans,
                    on=['Région', 'Céréales', 'Année'],
                    how='inner')

print(df_final1.head())
```

```python
# Fonction pour calculer le rendement avec un if
def calcul_rendement(row):
    if row['Superficie (en ha)'] > 0 and row['Production (en tonnes)'] > 0:
        return round(row['Production (en tonnes)'] / row['Superficie (en ha)'], 3)
    elif row['Superficie (en ha)'] >= 0 and row['Production (en tonnes)'] == 0:
        return 0
    else:
        return 'N.A.'

# Appliquer la fonction ligne par ligne
df_final1['Rendement (tonne/ha)'] = df_final1.apply(calcul_rendement, axis=1)


print(df_final1.head())
```

```python
# Transformer la base precipitation: années en colonne
df_precip_trans = pd.melt(df_precip,
                         id_vars=['Région'],  # on garde la colonne Région
                         var_name='Année',
                         value_name='Précipitations moyennes annuelles (en mm)')

# Nettoyer colonne Année
df_precip_trans = df_precip_trans[df_precip_trans['Année'].str.isnumeric()]
df_precip_trans['Année'] = df_precip_trans['Année'].astype(int)

# Trier par Région et Année
df_precip_trans = df_precip_trans.sort_values(by=['Région', 'Année']).reset_index(drop=True)

# Fusionner avec df.final1
df_final2 = pd.merge(df_final1,
                           df_precip_trans,
                           on=['Région', 'Année'],
                           how='left')  # on conserve toutes les lignes de df.final

# Trier aussi la base finale
df_final2 = df_final2.sort_values(by=['Région', 'Céréales', 'Année']).reset_index(drop=True)
```

```python
# Transformer la base jour de pluies: années en colonne
df_pluie_trans = pd.melt(df_pluie,
                         id_vars=['Région'],  # on garde la colonne Région
                         var_name='Année',
                         value_name='Nombre de jours de pluie ')

# Nettoyer colonne Année
df_pluie_trans = df_pluie_trans[df_pluie_trans['Année'].str.isnumeric()]
df_pluie_trans['Année'] = df_pluie_trans['Année'].astype(int)

# Trier par Région et Année
df_pluie_trans = df_pluie_trans.sort_values(by=['Région', 'Année']).reset_index(drop=True)

# Fusionner avec df.final2
df_final3 = pd.merge(df_final2,
                           df_pluie_trans,
                           on=['Région', 'Année'],
                           how='left')  # on conserve toutes les lignes de df.final2

# Trier aussi la base finale
df_final3 = df_final3.sort_values(by=['Région', 'Céréales', 'Année']).reset_index(drop=True)
```
